Q2. Computing duration¶
Now let's compute the duration variable. It should contain the duration of a ride in minutes.

What's the standard deviation of the trips duration in January?
* 42.59 

```python

# Ensure datetime columns are in datetime format
data_01['tpep_pickup_datetime'] = pd.to_datetime(data_01['tpep_pickup_datetime'])
data_01['tpep_dropoff_datetime'] = pd.to_datetime(data_01['tpep_dropoff_datetime'])

# Compute duration in minutes
data_01['duration'] = (data_01['tpep_dropoff_datetime'] - data_01['tpep_pickup_datetime']).dt.total_seconds() / 60

# Calculate standard deviation of duration
std_duration = data_01['duration'].std()
print(f"Standard deviation of trip duration in January: {std_duration:.2f} minutes")
```

Q3. Dropping outliers¶
Next, we need to check the distribution of the duration variable. There are some outliers. Let's remove them and keep only the records where the duration was between 1 and 60 minutes (inclusive).

What fraction of the records left after you dropped the outliers?

* 98%

```python

# Filter for durations between 1 and 60 minutes (inclusive)
mask = (data_01['duration'] >= 1) & (data_01['duration'] <= 60)
filtered_data_01 = data_01[mask]

# Calculate the fraction of records left
fraction_left = len(filtered_data_01) / len(data_01)
print(f"Fraction of records left after removing outliers: {fraction_left:.3f}")
```

Q4. One-hot encoding¶
Let's apply one-hot encoding to the pickup and dropoff location IDs. We'll use only these two features for our model.

Turn the dataframe into a list of dictionaries (remember to re-cast the ids to strings - otherwise it will label encode them)
Fit a dictionary vectorizer
Get a feature matrix from it
What's the dimensionality of this matrix (number of columns)?

* 155

```python

# Select relevant columns and cast to string
categorical = ['PULocationID', 'DOLocationID']
filtered_data_01[categorical] = filtered_data_01[categorical].astype(str)

# Turn dataframe into a list of dictionaries
dicts = filtered_data_01[categorical].to_dict(orient='records')

# Fit DictVectorizer
dv = DictVectorizer()
X_train = dv.fit_transform(dicts)

# Check dimensionality
print(f"Feature matrix shape: {X_train.shape}")
```

Q5. Training a model¶
Now let's use the feature matrix from the previous step to train a model.

Train a plain linear regression model with default parameters, where duration is the response variable
Calculate the RMSE of the model on the training data
What's the RMSE on train?
* 7.64

```python

# Prepare target variable
y_train = filtered_data_01['duration'].values

# Train linear regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict on training data
y_pred = lr.predict(X_train)

# Calculate RMSE
rmse = mean_squared_error(y_train, y_pred, squared=False)
print(f"RMSE on train: {rmse:.2f}")
```

Q6. Evaluating the model¶
Now let's apply this model to the validation dataset (February 2023).

What's the RMSE on validation?
* 7.81

```python

# Prepare validation data
data_02['tpep_pickup_datetime'] = pd.to_datetime(data_02['tpep_pickup_datetime'])
data_02['tpep_dropoff_datetime'] = pd.to_datetime(data_02['tpep_dropoff_datetime'])
data_02['duration'] = (data_02['tpep_dropoff_datetime'] - data_02['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filter for durations between 1 and 60 minutes (inclusive)
mask_val = (data_02['duration'] >= 1) & (data_02['duration'] <= 60)
filtered_data_02 = data_02[mask_val]

# Prepare features
filtered_data_02[categorical] = filtered_data_02[categorical].astype(str)
dicts_val = filtered_data_02[categorical].to_dict(orient='records')
X_val = dv.transform(dicts_val)

# Prepare target
y_val = filtered_data_02['duration'].values

# Predict and calculate RMSE
y_pred_val = lr.predict(X_val)
rmse_val = mean_squared_error(y_val, y_pred_val, squared=False)
print(f"RMSE on validation: {rmse_val:.2f}")
```
